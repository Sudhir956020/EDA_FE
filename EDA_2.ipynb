```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'dsa' is your DataFrame
subset_df = dsa[dsa.Category == 'PRODUCTIVITY'][:5]

# Create a larger figure
plt.figure(figsize=(10, 6))

# Create the bar plot
ax = sns.barplot(y='App', x='Installs', data=subset_df)

# Show the values on top of the bars
for p in ax.patches:
    ax.annotate(format(p.get_width(), ',.0f'),
                (p.get_width(), p.get_y() + p.get_height() / 2),
                ha='center', va='center',
                xytext=(5, 0),  # Distance from the bar
                textcoords='offset points')

# Set plot labels and title
plt.xlabel('Installs')
plt.ylabel('App')
plt.title('Top 5 PRODUCTIVITY Apps by Installs')

# Show the plot
plt.show()
```
